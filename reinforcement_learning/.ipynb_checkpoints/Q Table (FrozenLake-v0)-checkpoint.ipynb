# Q Table Naive
This example is from https://medium.com/emergent-future/simple-reinforcement-learning-with-tensorflow-part-0-q-learning-with-tables-and-neural-networks-d195264329d0

```python
import gym
import numpy as np
```

```python
env = gym.make('FrozenLake-v0')
```

```python
# initialize a zero matrix with Q-Table size
Q = np.zeros([env.observation_space.n, env.action_space.n])
```

```python
lr = .85
dis_fac = .99
num_episodes = 5000

# List for total rewards and steps per episode
rList = []
for i in range(num_episodes):
    # reset env and get first new observation
    s = env.reset()
    rAll = 0
    d = False
    j = 0
    # Q-table learning algorithm
    while j < 99:
        j +=1
        # Choose an action by greedily picking from Q-table
        a = np.argmax(Q[s, :] + np.random.randn(1, env.action_space.n)*(1./(i+1)))
        # get new state and reward from env
        s1, r, d , _ = env.step(a)
        # update Q-table with new knowledge
        Q[s,a] = Q[s,a] + lr*(r + dis_fac*np.max(Q[s1,:]) - Q[s,a])
        rAll += r
        s = s1
        if d == True:
            break
    rList.append(rAll)
```

```python
rList
```

```python
print( 'Score over time: ' + str(sum(rList)/ num_episodes))
```

```python
print('The final Q-table is: ')
print(Q)
```

```python
env.close()
```
